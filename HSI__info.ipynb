```python
import spectral.io.envi as envi
import numpy as np
import matplotlib.pyplot as plt


hdr_file_path = "/content/drive/My Drive/MHSI/030406-20x-roi1.hdr"


hsi = envi.open(hdr_file_path)


print("Metadata:", hsi.metadata)

hsi_data = hsi.load()
print(f"HSI Shape: {hsi_data.shape}")  # Example: (256, 320, 100) -> 100 spectral bands

num_bands = hsi_data.shape[2]
print(f"Number of Spectral Bands: {num_bands}")

print(f"Data Type: {hsi_data.dtype}")
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
for i, band in enumerate([0, num_bands//2, num_bands-1]):
    axes[i].imshow(hsi_data[:, :, band], cmap='gray')
    axes[i].set_title(f"Band {band + 1}")
    axes[i].axis('off')
plt.show()
```
